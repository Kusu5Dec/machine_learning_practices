```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#scaling (optional)>> brings all the feature on the same scale
```

```python
#why sacling/-------- To reduce the compution time(faster optimisation) and better interpretation
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
# standardisation >> zscore= X-mu/sigma

mean=np.mean(df['total_bill'])
```

```python
std=np.mean(df['total_bill'])
```

```python
std_data=[]

for i in list(df['total_bill']):
    zscore=(i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
#  scalling is optional because it doesnt change the distrubtion of data
# before scaling

sns.distplot(df[['total_bill']])
```

```python
#after scaling
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler

# Create the scaler object
scaler = StandardScaler()

# Assuming df is your DataFrame and 'total_bill' is the column you want to scale
# Reshape the data because fit() expects a 2D array
scaler.fit(df[['total_bill']])

```

```python
scaler.transform(df[['total_bill']])
```
