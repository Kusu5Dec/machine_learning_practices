```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=1)
```

```python
y     #in unsl you will not use
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test,=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=3)
```

```python
Kmeans
```

```python
Kmeans.fit(X_train)  #fit predict
```

```python
y_labels=Kmeans.predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
# how better the clustering model is 

from sklearn.metrics import silhouette_score
silhouette_score(X_train,Kmeans.labels_)
```

```python
Kmeans.labels_
```

```python
Kmeans.inertia_  #for k=3 you have wcss distance
```

```python
wcss = []
for k in range(1,10):
    Kmeans=KMeans(n_clusters=k)
    Kmeans.fit(X_train)
    wcss.append(Kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xticks(range(1,10))
plt.xlabel("No of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
# based on the above plot the optimal no of k cluster is 3
```

```python
from kneed import DataGenerator, KneeLocator
k1=KneeLocator(range(1,10),wcss,curve='convex',direction='decreasing')
```

```python
silhouette_coeff=[]
for k in range(2,10):
    Kmean=KMeans(n_clusters=k)
    Kmean.fit(X_train)
    score=silhouette_score(X_train,Kmeans.labels_)
    silhouette_coeff.append(score)
```
